```python
from tkinter import *
from tkinter.messagebox import *
import random
import numpy

class Morpion :
    def __init__(self) :
        self.C1RC = 0
        self.C1R = 0
        self.C2RC = 0
        self.C2R = 0
        self.C3RC = 0
        self.C3R = 0
        self.L1RC = 0
        self.L1R = 0
        self.L2RC = 0
        self.L2R = 0
        self.L3RC = 0
        self.L3R = 0
        self.the_end = False
        self.X = 0
        self.Y = 0
        self.a=1

        self.C1=[2,2,2]
        self.L1=[2,2,2]

        self.C2=[3,3,3]
        self.L2=[3,3,3]

        self.C3=[4,4,4]
        self.L3=[4,4,4]
        self.player_move = [
            [0,0,0],
            [0,0,0],
            [0,0,0]
        ]
        self.circle_position = [
            [(50,50),(150,50),(250,50)],
            [(50,150),(150,150),(250,150)],
            [(50,250),(150,250),(250,250)]
        ]
    
    def Clic(self, event):
        # position du pointeur de la souris
        self.X = event.x
        self.Y = event.y

        #Si self.a=1 on met une croix
        if self.a==1:
            if self.X < 100:
                if self.Y < 100:
                    if self.C1[0]==2:
                        self.a=0
                        self.C1[0]=1
                        self.L1[0]=1
                        self.Croix(50,50)
                    else :
                            showinfo(message='You can\'t !')
                else :
                    if self.Y < 200:
                        if self.C1[1]==2:
                            self.a=0
                            self.C1[1]=1
                            self.L2[0]=1
                            self.Croix(50,150)
                        else :
                            showinfo(message='You can\'t !')
                    else :
                        if self.C1[2]==2:
                            self.a=0
                            self.C1[2]=1
                            self.L3[0]=1
                            self.Croix(50,250)
                        else :
                            showinfo(message='You can\'t !')

            if self.X < 200 and self.X > 100:
                if self.Y < 100:
                    if self.C2[0]==3:
                        self.a=0
                        self.C2[0]=1
                        self.L1[1]=1
                        self.Croix(150,50)
                    else :
                            showinfo(message='You can\'t !')
                else :
                    if self.Y < 200:
                        if self.C2[1]==3:
                            self.a=0
                            self.C2[1]=1
                            self.L2[1]=1
                            self.Croix(150,150)
                        else :
                            showinfo(message='You can\'t !')
                    else :
                        if self.C2[2]==3:
                            self.a=0
                            self.C2[2]=1
                            self.L3[1]=1
                            self.Croix(150,250)
                        else :
                            showinfo(message='You can\'t !')

            if self.X < 300 and self.X > 200:
                if self.Y < 100:
                    if self.C3[0]==4:
                        self.a=0
                        self.C3[0]=1
                        self.L1[2]=1
                        self.Croix(250,50)
                    else :
                            showinfo(message='You can\'t !')

                else :
                    if self.Y < 200:
                        if self.C3[1]==4:
                            self.a=0
                            self.C3[1]=1
                            self.L2[2]=1
                            self.Croix(250,150)
                        else :
                            showinfo(message='You can\'t !')

                    else :
                        if self.C3[2]==4:
                            self.a=0
                            self.C3[2]=1
                            self.L3[2]=1
                            self.Croix(250,250)
                        else :
                            showinfo(message='You can\'t !')
        if self.the_end is False:
            #pour switch entre les ia 
            #self.IaDumb(X,Y)
            self.IaSmart()
            
    def IaSmart_draw(self,column_index, ligne_index):
        self.a=1
        exec("self.C" + str(column_index+1) + "[" + str(ligne_index) + "]" + " = " + "0")
        exec("self.L" + str(ligne_index+1) + "[" + str(column_index) + "]" + " = " + "0")
        self.Rond(self.circle_position[ligne_index][column_index][0],self.circle_position[ligne_index][column_index][1])
        self.player_move[ligne_index][column_index] = -1
        self.check_is_finish = True
        
    def IaSmart(self):
        print("joueur 1", self.X, self.Y)
        self.check_is_finish = False
        #vérifier où le joueur joue
        if self.X < 100:
            if self.Y < 100:
                self.player_move[0][0] = 1

            else :
                if self.Y < 200:
                    self.player_move[1][0] = 1
                else :
                    self.player_move[2][0] = 1
        if self.X < 200 and self.X > 100:
            if self.Y < 100:
                self.player_move[0][1] = 1
            else :
                if self.Y < 200:
                    self.player_move[1][1] = 1

                else :
                    self.player_move[2][1] = 1

        if self.X < 300 and self.X > 200:
            if self.Y < 100:
                self.player_move[0][2] = 1
            else :
                if self.Y < 200:
                    self.player_move[1][2] = 1

                else :
                    self.player_move[2][2] = 1
        self.player_move_diagonal = [
            [
                self.player_move[0][0],
                self.player_move[1][1],
                self.player_move[2][2]
            ],
            [
                self.player_move[0][2],
                self.player_move[1][1],
                self.player_move[2][0]
            ],
            
        ]
        total = 0
        for p_index,p in enumerate(self.player_move):
            sum_p = sum(p)
            print(p)
            if 1 in p :
                total +=1
            #vérifier si la ligne est presque validée
            if sum_p == 2 :
                ligne_index = p_index
                column_index = self.player_move[ligne_index].index(0)
                self.IaSmart_draw(column_index, ligne_index)
                print("bloquage ligne")
                break
            #vérifier si c'est le premier mouvement ou si ne sais pas 
            if (total == 1 and p_index == len(self.player_move)-1) or(total > 1 and p_index == len(self.player_move)-1):
                ligne_index = random.randrange(3)
                if 0 in self.player_move[ligne_index]:
                    column_index = self.player_move[ligne_index].index(0)
                    self.IaSmart_draw(column_index, ligne_index)
                    print("premier move ou ne sais pas")
                    break
            #vérifier si la colonne est presque validée
            column = [row[p_index] for row in self.player_move]
            sum_column = sum(column)
            if sum_column == 2 :
                if 0 in column:
                    ligne_index = column.index(0)
                    column_index = p_index
                    self.IaSmart_draw(column_index, ligne_index)
                    print("bloquage colonne")
                    print(column)
                    break
            
            #vérifier si la diagonale est presque validée   
            for d_index,d in enumerate(self.player_move_diagonal):
                sum_diagonal = sum(d)
                print(d, sum_diagonal, "diagonale")
                if (sum_diagonal == 2) and (0 in d):
                    ligne_index = d.index(0)
                    column_index = ligne_index if d_index == 0 else (len(d)-1) - ligne_index
                    self.IaSmart_draw(column_index, ligne_index)
                    print("bloquage  diagonale")
                    print(column)
                    break
                
            if self.check_is_finish:
                break
            
                        
                    
            
    def IaDumb(self,X,Y):
        if self.a == 0:
            self.X = random.randint(0,300)
            self.X = random.randint(0,300)
            if self.X < 100:
                if self.Y < 100:
                    if self.C1[0]==2:
                        self.a=1
                        self.C1[0]=0
                        self.L1[0]=0
                        self.Rond(50,50)
                    else :
                        self.IaDumb(X,Y)

                else :
                    if self.Y < 200:
                        if self.C1[1]==2:
                            self.a=1
                            self.C1[1]=0
                            self.L2[0]=0
                            self.Rond(50,150)
                        else :
                            self.IaDumb(X,Y)

                    else :
                        if self.C1[2]==2:
                            self.a=1
                            self.C1[2]=0
                            self.L3[0]=0
                            self.Rond(50,250)
                        else :
                            self.IaDumb(X,Y)

            if self.X < 200 and self.X > 100:
                if self.Y < 100:
                    if self.C2[0]==3:
                        self.a=1
                        self.C2[0]=0
                        self.L1[1]=0
                        self.Rond(150,50)
                    else :
                        self.IaDumb(X,Y)

                else :
                    if self.Y < 200:
                        if self.C2[1]==3:
                            self.a=1
                            self.C2[1]=0
                            self.L2[1]=0
                            self.Rond(150,150)
                        else :
                            self.IaDumb(X,Y)

                    else :
                        if self.C2[2]==3:
                            self.a=1
                            self.C2[2]=0
                            self.L3[1]=0
                            self.Rond(150,250)
                        else :
                            self.IaDumb(X,Y)

            if self.X < 300 and self.X > 200:
                if self.Y < 100:
                    if self.C3[0]==4:
                        self.a=1
                        self.C3[0]=0
                        self.L1[2]=0
                        self.Rond(250,50)
                    else :
                        self.IaDumb(X,Y)
                else :
                    if self.Y < 200:
                        if self.C3[1]==4:
                            self.a=1
                            self.C3[1]=0
                            self.L2[2]=0
                            self.Rond(250,150)
                        else :
                            self.IaDumb(X,Y)

                    else :
                        if self.C3[2]==4:
                            self.a=1
                            self.C3[2]=0
                            self.L3[2]=0
                            self.Rond(250,250)
                        else :
                            self.IaDumb(X,Y)
                
    #ici on créer la fonction pour afficher les ronds
    def Rond(self, x1,y1):
        Canevas.create_oval(x1-45,y1-45,x1+45,y1+45)
        self.Verif()

    #ici on créer la fonction pour afficher les croix
    def Croix(self, x1,y1):
        Canevas.create_line(x1-45,y1-45,x1+45,y1+45)
        Canevas.create_line(x1+45,y1-45,x1-45,y1+45)
        self.Verif()

    #ici on créer la fonction qui compte les points
    def Verif(self):
        self.C1RC = self.C1.count(1)
        self.C1R = self.C1.count(0)
        self.C2RC = self.C2.count(1)
        self.C2R = self.C2.count(0)
        self.C3RC = self.C3.count(1)
        self.C3R = self.C3.count(0)

        self.L1RC = self.L1.count(1)
        self.L1R = self.L1.count(0)
        self.L2RC = self.L2.count(1)
        self.L2R = self.L2.count(0)
        self.L3RC = self.L3.count(1)
        self.L3R = self.L3.count(0)


        #ici on affiche le joueur qui a gagné
        self.the_end = False
        if self.C1RC == 3:
            self.the_end = True
            showinfo(title='End of the game',message='Player 1 won !')
        if self.C1R == 3:
            showinfo(title='End of the game',message='Player 2 won !')
        if self.C2RC == 3:
            self.the_end = True
            showinfo(title='End of the game',message='Player 1 won !')
        if self.C2R == 3:
            showinfo(title='End of the game',message='Player 2 won !')
        if self.C3RC == 3:
            self.the_end = True
            showinfo(title='End of the game',message='Player 1 won !')
        if self.C3R == 3:
            showinfo(title='End of the game',message='Player 2 won !')

        if self.C1[0]==1 and self.C2[1]==1 and self.C3[2]==1:
            self.the_end = True
            showinfo(title='End of the game',message='Player 1 won !')
        if self.C1[2]==1 and self.C2[1]==1 and self.C3[0]==1:
            self.the_end = True
            showinfo(title='End of the game',message='Player 1 won !')    

        if self.C1[0]==0 and self.C2[1]==0 and self.C3[2]==0:
            showinfo(title='End of the game',message='Player 2 won !')
        if self.C1[2]==0 and self.C2[1]==0 and self.C3[0]==0:
            showinfo(title='End of the game',message='Player 2 won !')  

        if self.L1RC == 3:
            self.the_end = True
            showinfo(title='End of the game',message='Player 1 won !')
        if self.L1R == 3:
            showinfo(title='End of the game',message='Player 2 won !')
        if self.L2RC == 3:
            self.the_end = True
            showinfo(title='End of the game',message='Player 1 won !')
        if self.L2R == 3:
            showinfo(title='End of the game',message='Player 2 won !')
        if self.L3RC == 3:
            self.the_end = True
            showinfo(title='End of the game',message='Player 1 won !')
        if self.L3R == 3:
            showinfo(title='End of the game',message='Player 2 won !')
   
       



# Création de la fenêtre principale
Mafenetre = Tk()
Mafenetre.title("Morpion")

# Création d'un widget Canvas
Largeur = 300
Hauteur = 300
Canevas = Canvas(Mafenetre, width = Largeur, height =Hauteur, bg ="white")

morpion = Morpion()
# La méthode bind() permet de lier un événement avec une fonction :
# un clic sur la zone graphique provoquera l'appel de la fonction utilisateur Clic()
Canevas.bind("<Button-1>", morpion.Clic)
Canevas.pack(padx =5, pady =5)

#ici on créer les lignes qui délimite les colones et les cases
Canevas.create_line(0,100,300,100,fill="purple",width=4)

Canevas.create_line(0,200,300,200,fill="purple",width=4)

Canevas.create_line(100,300,300,-100000,fill="purple",width=4)

Canevas.create_line(200,300,300,-100000,fill="purple",width=4)


# Création d'un widget Button (bouton Quitter)
Button(Mafenetre, text ="Left", command = Mafenetre.destroy).pack(side=LEFT,padx=5,pady=5)

Mafenetre.mainloop()
```
